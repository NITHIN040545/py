```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
from sklearn.decomposition import PCA
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("Mall_Customers.csv")
```

```python
df.head()
```

```python
def preprocess_data(df, features):
    df_filtered = df.dropna(subset=features)
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(df_filtered[features])
    return X_scaled, df_filtered.index, scaler
```

```python
# Elbow method for choosing optimal K
def plot_elbow(X_scaled):
    inertia = []
    k_range = range(2, 11)

    for k in k_range:
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(X_scaled)
        inertia.append(kmeans.inertia_)

    plt.figure(figsize=(10, 6))
    plt.plot(k_range, inertia, marker='o')
    plt.title('Elbow Method for Optimal K')
    plt.xlabel('Number of Clusters (K)')
    plt.ylabel('Inertia')
    plt.grid()
    plt.xticks(k_range)
    plt.savefig("elbow_plot.png")
    plt.show()
```

```python

# Run KMeans
def run_kmeans(X_scaled, n_clusters):
    model = KMeans(n_clusters=n_clusters, random_state=42)
    labels = model.fit_predict(X_scaled)
    return labels, model
```

```python
# Evaluate clustering performance
def evaluate_clustering(X_scaled, labels):
    print("\nClustering Evaluation Metrics:")
    silhouette = silhouette_score(X_scaled, labels)
    calinski = calinski_harabasz_score(X_scaled, labels)
    davies = davies_bouldin_score(X_scaled, labels)

    print(f'Silhouette Score: {silhouette:.4f}')
    print(f'Calinski-Harabasz Index: {calinski:.4f}')
    print(f'Davies-Bouldin Index: {davies:.4f}')
```

```python
# Visualize clusters using PCA
def visualize_clusters(df, features, cluster_col):
    X = df[features]
    pca = PCA(n_components=2)
    components = pca.fit_transform(X)
    plot_df = pd.DataFrame(data=components, columns=['PC1', 'PC2'])
    plot_df[cluster_col] = df[cluster_col].astype(str)

    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=plot_df, x='PC1', y='PC2', hue=cluster_col, palette='tab10', s=60)
    plt.title('Cluster Visualization using PCA')
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.legend(title='Cluster')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
# Profile clusters
def cluster_profiling(df, features, cluster_col):
    for c in sorted(df[cluster_col].unique()):
        print(f"\nCluster {c} Summary Statistics:\n")
        print(df[df[cluster_col] == c][features].describe())
```

```python
# Main function pipeline
def main():
    features = ['Annual Income (k$)', 'Spending Score (1-100)']
    df = pd.read_csv("Mall_Customers.csv")

    X_scaled, valid_idx, scaler = preprocess_data(df, features)
    print('âœ… Data loaded and preprocessed.')

    print('\n--- ðŸ“‰ Elbow Method ---')
    plot_elbow(X_scaled)

    n_clusters = 4
    labels, kmeans_model = run_kmeans(X_scaled, n_clusters)
    df.loc[valid_idx, 'Cluster'] = labels

    evaluate_clustering(X_scaled, labels)

    print('\n--- ðŸ“Š Cluster Means ---')
    print(df.groupby('Cluster')[features].mean())

    print('\n--- ðŸŽ¨ Cluster Visualization ---')
    visualize_clusters(df.loc[valid_idx], features, 'Cluster')

    print('\n--- ðŸ“Œ Cluster Profiling ---')
    cluster_profiling(df.loc[valid_idx], features, 'Cluster')


# Run the full pipeline
main()
```
